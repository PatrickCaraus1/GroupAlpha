```python
import sys
import os
import pandas as pd
import kagglehub
```

```python

# Download latest version
path = kagglehub.dataset_download("shubhamchandra235/imdb-and-tmdb-movie-metadata-big-dataset-1m", output_dir="./data")

print("Path to dataset files:", path)
```

```python
# files = os.listdir("IMDB TMDB Movie Metadata Big Dataset (1M).csv")
# print("Files in dataset:", files)
```

```python
movies_file = os.path.join(path, "IMDB TMDB Movie Metadata Big Dataset (1M).csv")
movies = pd.read_csv(movies_file)
print(movies.head())
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)

movies.describe()
```

```python

print("Highest rating:", movies["IMDB_Rating"].max())
print("Lowest rating:", movies["IMDB_Rating"].min())
print("Average rating:", movies["IMDB_Rating"].mean())
print("Median rating:", movies["IMDB_Rating"].median())
print("Standard deviation:", movies["IMDB_Rating"].std())
```
